```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Carregar os dados
df = pd.read_csv("C:\\Users\\marco\\OneDrive\\rea de Trabalho\\Hackathon\\digital-twin-gas\\backend\\data\\synthetic\\generated\\dataset.csv", parse_dates=['timestamp'])

# Visualizar as primeiras linhas
print(df.head())

# Verificar informaes bsicas
print(df.info())

# Estatsticas descritivas
print(df.describe())

# Verificar balanceamento das classes
print(df['leak_label'].value_counts())

# Plotar distribuio das classes
sns.countplot(x='leak_label', data=df)
plt.title('Distribuio de Vazamentos')
plt.show()

# Correlao entre features
plt.figure(figsize=(10, 6))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm')
plt.title('Matriz de Correlao')
plt.show()
```
